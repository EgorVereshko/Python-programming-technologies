Как видим, не очень удобно, что возвращаются кортежи. Удобней было бы, чтобы словари.

```python
c.description # вот тут есть описания полей.
```

```python
def dict_factory(cursor, row): 
    # обертка для преобразования 
    # полученной строки. (взята из документации)
    d = {}
    for idx, col in enumerate(cursor.description):
        d[col[0]] = row[idx]
    return d
```

```python
conn.row_factory = dict_factory
```

```python
с = conn.cursor()
c.execute("SELECT * FROM netflix_titles limit 3")
c.fetchall() # Теперь буду словари
```

```python
c.execute("SELECT count(*) as cnt FROM netflix_titles where listed_in='Stand-Up Comedy'")
c.fetchall() # Сколько вообще стендап комедий
```

```python
c.execute("SELECT rating, count(*) as cnt FROM netflix_titles group by rating")
c.fetchall() # Группируем по рейтингу
# Важно тут сказать, что SQL -- замена pandas полноценная, 
# и всю обработку данных можно делать прямо внутри него
```

```python
c.execute("ALTER TABLE netflix_titles RENAME COLUMN cast TO actors;")
```

```python
with conn:
    c.execute("""
        CREATE TABLE actors_list (
            name TEXT NOT NULL,
            show_id INT,
            FOREIGN KEY(show_id) REFERENCES netflix_titles(show_id)
        )
""")
```

```python
table = c.execute("SELECT show_id, actors FROM netflix_titles").fetchall()
```

```python
for row in table:
    show_id = row[0]
    act = row[1].split(", ")
    data = []
    print(show_id, act)
    for name in act:
        data.append((name, int(show_id)))
```

```python
with conn:
    c.executemany("INSERT INTO actors_list VALUES(?, ?)", data)
```

```python
with conn:
    c.execute("""
        CREATE TABLE films_list (
            show_id INT,
            film_name TEXT NOT NULL
        )
""")
```

```python
table_films = c.execute("SELECT show_id, title FROM netflix_titles WHERE type='Movie'").fetchall()
```

```python
with conn:
    for row in table_films:
        c.execute("INSERT INTO films_list VALUES(?, ?)", (row[1], int(row[0])))
```

```python
with conn:
     c.execute("""
            CREATE TABLE table_act_films AS
            SELECT actors_list.name, actors_list.show_id, films_list.film_name
            FROM actors_list
            INNER JOIN films_list
            ON actors_list.show_id = films_list.show_id;
        );
""")
```

```python
with conn:
    c.execute("""
        SELECT actors_list.name AS actor1, films_list.name AS actor2, COUNT(*) AS count
        FROM actors_list
        JOIN films_list
        ON actors_list.film_name = films_list.film_name AND actors_list.name < films_list.name
        GROUP BY actor1, actor2
        ORDER BY count DESC LIMIT 1;
""")
```
